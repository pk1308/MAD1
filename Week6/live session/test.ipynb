```python
import json

def is_valid_json(json_string):
  try:
      json.loads(json_string)
      return True
  except ValueError:
      return False

```

```python
data =  {
     "name": "Harvard",
     "brands": ("Rolex", "Porsche"),
     "max" :42
   }
```

```python
is_valid_json(  json.dump(data))
```

```python
is_valid_json(      {
     "Cities": ["Mumbai", "Chennai"],
     "stable": True,
     "Objects":"jj"
   })
```
